**Price an ATM European call and put using a binomial tree:**

- Choose the number of steps in the tree you see convenient to achieve
reliable estimates.

```python
import math
import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.linear_model import LinearRegression   # for LSM regression
```

```python
S0 = 100.0
K = 100.0  # ATM
r = 0.05   # annual risk-free rate
T = 0.25   # 3 months = 0.25 years
sigma1 = 0.20
sigma2 = 0.25  # for vega comparison (5% absolute increase)

def binomial_european_option(S0, K, r, T, sigma, N=200, option_type='call'):
    dt = T / N
    u = math.exp(sigma * math.sqrt(dt))
    d = 1.0 / u
    disc = math.exp(-r * dt)
    p = (math.exp(r * dt) - d) / (u - d)
    # terminal prices
    prices = [S0 * (u**j) * (d**(N-j)) for j in range(N+1)]
    if option_type == 'call':
        values = [max(pn - K, 0.0) for pn in prices]
    else:
        values = [max(K - pn, 0.0) for pn in prices]

    # backward induction
    for i in range(N-1, -1, -1):
        values = [disc * (p * values[j+1] + (1-p) * values[j]) for j in range(i+1)]
    return values[0], u, d, p, dt

# Run with two choices of N to show convergence
results = []
for N in [50, 100, 200]:
    C, u, d, p, dt = binomial_european_option(S0, K, r, T, sigma1, N=N, option_type='call')
    P, _, _, _, _ = binomial_european_option(S0, K, r, T, sigma1, N=N, option_type='put')
    results.append({'N': N, 'Call Price': C, 'Put Price': P})

formatted_results = []
for item in results:
    formatted_results.append({
        'N': item['N'],
        'Call Price': f"{item['Call Price']:.3f}",
        'Put Price': f"{item['Put Price']:.3f}"
    })

pprint.pprint(formatted_results)
```

```python
def compute_delta(S0, K, r, T, sigma, N=200):
    # compute option price at time 0 and first node values to estimate delta
    C0, u, d, p, dt = binomial_european_option(S0, K, r, T, sigma, N=N, option_type='call')
    P0, _, _, _, _ = binomial_european_option(S0, K, r, T, sigma, N=N, option_type='put')
    # One-step ahead option values (value at node up and down) using N steps but inspect first iteration
    # We'll compute option values at t=dt using a tree of N steps by evaluating continuation at first nodes.
    # Simpler: compute option prices with N=1 for first-step up/down for delta estimate using same params.
    Cu, _, _, _, _ = binomial_european_option(S0*u, K, r, T-dt, sigma, N=1, option_type='call')
    Cd, _, _, _, _ = binomial_european_option(S0*d, K, r, T-dt, sigma, N=1, option_type='call')
    Pu, _, _, _, _ = binomial_european_option(S0*u, K, r, T-dt, sigma, N=1, option_type='put')
    Pd, _, _, _, _ = binomial_european_option(S0*d, K, r, T-dt, sigma, N=1, option_type='put')
    # delta approximations
    delta_call = (Cu - Cd) / (S0*u - S0*d)
    delta_put  = (Pu - Pd) / (S0*u - S0*d)
    return C0, P0, delta_call, delta_put
```

```python
def binomial_american_option(S, K, T, r, sigma, N, option_type='call'):
    dt = T / N
    disc = np.exp(-r * dt)
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    # Initialize asset prices at maturity
    asset_prices = np.zeros(N + 1)
    for i in range(N + 1):
        asset_prices[i] = S * (u ** (N - i)) * (d ** i)

    # Initialize option values at maturity
    option_values = np.zeros(N + 1)
    if option_type == 'call':
        option_values = np.maximum(0, asset_prices - K)
    else:
        option_values = np.maximum(0, K - asset_prices)

    # Backward induction for American options
    for j in range(N - 1, -1, -1):
        for i in range(j + 1):
            option_values[i] = disc * (p * option_values[i] + (1 - p) * option_values[i + 1])
            # Check for early exercise
            if option_type == 'call':
                option_values[i] = max(option_values[i], asset_prices[i] - K)
            else:
                option_values[i] = max(option_values[i], K - asset_prices[i])

    return option_values[0]

# Parameters
S = 100  # Underlying stock price
K = 100  # Strike price (ATM)
T = 3 / 12  # Time to expiration in years
r = 0.05  # Annual risk-free rate
sigma = 0.2  # Volatility (assumption)

N_values = [25, 50, 100, 200]
results = {}

for N in N_values:
    call_price = binomial_american_option(S, K, T, r, sigma, N, option_type='call')
    put_price = binomial_american_option(S, K, T, r, sigma, N, option_type='put')
    results[N] = {'Call Price': call_price, 'Put Price': put_price}

# Print results
for N in results:
    print(f"N = {N}: Call Price = {results[N]['Call Price']:.4f}, Put Price = {results[N]['Put Price']:.4f}")
```

```python
def binomial_american_option(S, K, T, r, sigma, N, option_type='call'):
    dt = T / N
    discount_factor = np.exp(-r * dt)
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    # Initialize asset prices at maturity
    asset_prices = np.zeros(N + 1)
    for i in range(N + 1):
        asset_prices[i] = S * (u ** (N - i)) * (d ** i)

    # Initialize option values at maturity
    option_values = np.zeros(N + 1)
    if option_type == 'call':
        option_values = np.maximum(0, asset_prices - K)
    else:
        option_values = np.maximum(0, K - asset_prices)

    # Backward induction for American options
    for j in range(N - 1, -1, -1):
        for i in range(j + 1):
            option_values[i] = discount_factor * (p * option_values[i] + (1 - p) * option_values[i + 1])
            if option_type == 'call':
                option_values[i] = max(option_values[i], asset_prices[i] - K)
            else:
                option_values[i] = max(option_values[i], K - asset_prices[i])

    return option_values[0]  # Return the option value at the root node

# Parameters
S = 100  # Underlying stock price
K = 100  # Strike price (ATM)
T = 3 / 12  # Time to expiration in years
r = 0.05  # Annual risk-free rate
sigma_initial = 0.20  # Initial volatility at 20%
sigma_increase = 0.25  # Increased volatility at 25%

N_values = [25, 50, 100,200]
results = {}

for N in N_values:
    # Calculate prices with initial volatility
    call_price_initial = binomial_american_option(S, K, T, r, sigma_initial, N, option_type='call')
    put_price_initial = binomial_american_option(S, K, T, r, sigma_initial, N, option_type='put')

    # Calculate prices with increased volatility
    call_price_increased = binomial_american_option(S, K, T, r, sigma_increase, N, option_type='call')
    put_price_increased = binomial_american_option(S, K, T, r, sigma_increase, N, option_type='put')

    # Compute the change in prices
    call_price_change = call_price_increased - call_price_initial
    put_price_change = put_price_increased - put_price_initial

    results[N] = {
        'Call Price Change': call_price_change,
        'Put Price Change': put_price_change
    }

# Print results
for N in results:
    print(f"N = {N}: Call Price Change = {results[N]['Call Price Change']:.4f}, Put Price Change = {results[N]['Put Price Change']:.4f}")
```

```python
# Parameters
S0 = 100.0
r = 0.05
sigma = 0.20
T = 3/12
N = 200
dt = T / N

def trinomial_params(r, sigma, dt):
    u = math.exp(sigma * math.sqrt(3 * dt))
    d = 1.0 / u
    a = (sigma**2 * dt + r * dt) / (sigma * math.sqrt(3 * dt))
    p_u = (a**2 + a)/6.0
    p_d = (a**2 - a)/6.0
    p_m = 1.0 - (a**2)/3.0
    return u, d, p_u, p_m, p_d

u, d, p_u, p_m, p_d = trinomial_params(r, sigma, dt)

def trinomial_option_price(S0, K, r, sigma, T, N, option_type='call', american=False):
    dt = T / N
    u, d, p_u, p_m, p_d = trinomial_params(r, sigma, dt)
    discount = math.exp(-r * dt)

    # Terminal asset prices: j = -N..N mapped to 0..2N
    J = np.arange(-N, N+1)
    ST = S0 * (u ** J)
    if option_type == 'call':
        values = np.maximum(ST - K, 0.0)
    else:
        values = np.maximum(K - ST, 0.0)

    # Backward induction
    for i in range(N-1, -1, -1):
        new_values = np.zeros(2*i + 1)
        for idx, j in enumerate(range(-i, i+1)):
            idx_up = (j+1) + (i+1)
            idx_mid = j + (i+1)
            idx_dn = (j-1) + (i+1)
            cont = discount * (p_u * values[idx_up] + p_m * values[idx_mid] + p_d * values[idx_dn])
            if american:
                S = S0 * (u**j)
                if option_type == 'call':
                    exc = max(S - K, 0.0)
                else:
                    exc = max(K - S, 0.0)
                new_values[idx] = max(cont, exc)
            else:
                new_values[idx] = cont
        values = new_values
    return float(values[0])

# Strike list
strike_prices = [90, 95, 100, 105, 110]

# Compute prices
rows = []
for K in strike_prices:
    eur_put = trinomial_option_price(S0, K, r, sigma, T, N, option_type='put', american=False)
    eur_call = trinomial_option_price(S0, K, r, sigma, T, N, option_type='call', american=False)
    amer_put = trinomial_option_price(S0, K, r, sigma, T, N, option_type='put', american=True)
    amer_call = trinomial_option_price(S0, K, r, sigma, T, N, option_type='call', american=True)
    rows.append({
        'Strike': K,
        'Moneyness (K/S0)': K / S0,
        'European Put': eur_put,
        'American Put': amer_put,
        'European Call': eur_call,
        'American Call': amer_call
    })

df = pd.DataFrame(rows)


# Plots
plt.figure(figsize=(6,4))
plt.plot(df['Moneyness (K/S0)'], df['European Put'], marker='o', label='European Put')
plt.plot(df['Moneyness (K/S0)'], df['American Put'], marker='s', label='American Put')
plt.xlabel('Moneyness (K / S0)')
plt.ylabel('Put Price')
plt.title('Put Price vs Moneyness (European vs American)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(df['Moneyness (K/S0)'], df['European Call'], marker='o', label='European Call')
plt.plot(df['Moneyness (K/S0)'], df['American Call'], marker='s', label='American Call')
plt.xlabel('Moneyness (K / S0)')
plt.ylabel('Call Price')
plt.title('Call Price vs Moneyness (European vs American)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print("European & American Option Prices (Trinomial Tree)")
df
```

```python
#  Graph #2. Graph American call prices and put prices versus stock prices.
def american_binomial_option_price(S0, K, r, q, sigma, T, N, option_type='call'):
    # Cox-Ross-Rubinstein parameters
    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    pu = (np.exp((r - q) * dt) - d) / (u - d)
    pd = 1 - pu
    disc = np.exp(-r * dt)

    # Stock prices at maturity
    ST = np.array([S0 * (u ** j) * (d ** (N - j)) for j in range(N + 1)])
    if option_type == 'call':
        values = np.maximum(ST - K, 0.0)
    else:
        values = np.maximum(K - ST, 0.0)

    # Work backwards
    for i in range(N - 1, -1, -1):
        # move one step back: compute stock prices at this level
        ST = ST[:i + 1] * d  # adjust (previous ST were for level i+1)
        cont = disc * (pu * values[1:i + 2] + pd * values[0:i + 1])
        intrinsic = np.maximum(ST - K, 0.0) if option_type == 'call' else np.maximum(K - ST, 0.0)
        # American: price = max(continuation, intrinsic)
        values = np.maximum(cont, intrinsic)
    return values[0]

# Example parameters (change as needed)
K = 100.0
T = 0.5
r = 0.05
q = 0.0        # dividend yield (set >0 if underlying pays dividends)
sigma = 0.2
N = 200        # binomial steps (increase for accuracy)

# stock grid
S_grid = np.linspace(50, 150, 201)
calls = np.array([american_binomial_option_price(S, K, r, q, sigma, T, N, 'call') for S in S_grid])
puts  = np.array([american_binomial_option_price(S, K, r, q, sigma, T, N, 'put')  for S in S_grid])

# Plot
plt.figure(figsize=(9,6))
plt.plot(S_grid, calls, label='American Call')
plt.plot(S_grid, puts,  label='American Put')
plt.axvline(K, linestyle='--', label=f'Strike K={K}')
plt.xlabel('Stock price S'); plt.ylabel('Option price'); plt.title('American Call & Put vs S (Binomial)')
plt.legend(); plt.grid(True); plt.show()
```

```python
# Graph #3. Graph European and American call prices versus strike.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters (change to your assignment values)
S = 100.0      # current stock price (fixed)
T = 0.5
r = 0.05
q = 0.0        # dividend yield, set >0 if dividends exist
sigma = 0.2
N = 400        # binomial steps for American pricing

# Strike grid
K_grid = np.linspace(60,140,161)

# Black-Scholes European call (with dividend yield q)
def european_call_bs(S,K,r,q,sigma,T):
    if T <= 0:
        return max(S-K,0)
    d1 = (np.log(S/K) + (r - q + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return S*np.exp(-q*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

# CRR binomial American call
def american_binomial_call(S0,K,r,q,sigma,T,N):
    dt = T/N
    u = np.exp(sigma*np.sqrt(dt))
    d = 1/u
    pu = (np.exp((r - q)*dt) - d) / (u - d)
    disc = np.exp(-r*dt)
    # terminal stock prices
    ST = np.array([S0 * (u**j) * (d**(N-j)) for j in range(N+1)])
    values = np.maximum(ST - K, 0.0)
    for i in range(N-1, -1, -1):
        ST = ST[:i+1] * d
        cont = disc * (pu * values[1:i+2] + (1-pu) * values[0:i+1])
        intrinsic = np.maximum(ST - K, 0.0)
        values = np.maximum(cont, intrinsic)
    return values[0]

# Compute values across strikes
C_E = np.array([european_call_bs(S,K,r,q,sigma,T) for K in K_grid])
C_A = np.array([american_binomial_call(S,K,r,q,sigma,T,N) for K in K_grid])

# Plot
plt.figure(figsize=(9,6))
plt.plot(K_grid, C_E, label='European Call (BS)')
plt.plot(K_grid, C_A, label='American Call (Binomial)')
plt.axvline(S, color='k', linestyle='--', label=f'S = {S}')
plt.xlabel('Strike K'); plt.ylabel('Call price'); plt.title('European vs American Call Price vs Strike')
plt.legend(); plt.grid(True); plt.show()
```

```python
S0 = 180.0       # initial stock price
K = 182.0        # strike
r = 0.02         # continuous annual interest rate
sigma = 0.25     # volatility
T = 0.5          # time to maturity (years)
N = 25           # number of binomial steps

dt = T / N
u = math.exp(sigma * math.sqrt(dt))
d = 1.0 / u
discount = math.exp(-r * dt)
p = (math.exp(r * dt) - d) / (u - d)

params = {
    "S0": S0, "K": K, "r": r, "sigma": sigma, "T": T, "N": N,
    "dt": dt, "u": u, "d": d, "discount": discount, "p": p
}

params_df = pd.DataFrame(list(params.items()), columns=["parameter", "value"])
params_df
```

```python
# Build stock tree

stock = [[S0 * (u**i) * (d**(n-i)) for i in range(n+1)] for n in range(N+1)]
# Convert tree to a rectangular DataFrame for display
# Rows = time steps (0..N), Columns = node index i (0..N)

stock_table = pd.DataFrame(index=range(N+1), columns=[f"i={i}" for i in range(N+1)], dtype=float)
for n in range(N+1):
    for i in range(n+1):
        stock_table.at[n, f"i={i}"] = stock[n][i]
print("Stock price tree (S_{n,i})")

stock_table.round(2)
```

```python

# Price American put by backward induction

value = [[0.0]*(n+1) for n in range(N+1)]
# Terminal payoffs
for i in range(N+1):
    value[N][i] = max(K - stock[N][i], 0.0)

# Backward induction with early exercise check
for n in range(N-1, -1, -1):
    for i in range(n+1):
        cont = discount * (p * value[n+1][i+1] + (1-p) * value[n+1][i])
        exercise = max(K - stock[n][i], 0.0)
        value[n][i] = max(exercise, cont)

american_put_price = value[0][0]
print(f"American put price (N={N}): {american_put_price:.6f}")

# Convert value tree to DataFrame
value_table = pd.DataFrame(index=range(N+1), columns=[f"i={i}" for i in range(N+1)], dtype=float)
for n in range(N+1):
    for i in range(n+1):
        value_table.at[n, f"i={i}"] = value[n][i]
print("Option value tree (V_{n,i})")

value_table.round(2)

```

```python

# Compute deltas at each non-terminal node

delta = [[None]*(n+1) for n in range(N)]
delta_rows = []
for n in range(N):
    for i in range(n+1):
        v_up = value[n+1][i+1]
        v_down = value[n+1][i]
        s_up = stock[n+1][i+1]
        s_down = stock[n+1][i]
        dlt = (v_up - v_down) / (s_up - s_down)
        delta[n][i] = dlt
        delta_rows.append({"n": n, "i": i, "S": stock[n][i], "V": value[n][i], "delta_long": dlt})

delta_df = pd.DataFrame(delta_rows).round(8)
print("Delta matrix (flattened) - first 20 rows")

delta_df.head(20).round(2)


```

```python

#Choose hedging path

# Options: "alternating", "all_up", "all_down", "random"
path_kind = "alternating"
seed = 12345

if path_kind == "alternating":
    path = [1 if k % 2 == 0 else 0 for k in range(N)]
elif path_kind == "all_up":
    path = [1]*N
elif path_kind == "all_down":
    path = [0]*N
elif path_kind == "random":
    np.random.seed(seed)
    path = list((np.random.rand(N) < 0.5).astype(int))
else:
    raise ValueError("Unknown path kind")

# Display the chosen path (first 40 moves)
path_df = pd.DataFrame({"step": list(range(1, N+1)), "move(1=up,0=down)": path})
print(f"Chosen path ({path_kind}) - first 40 moves")
path_df.head(40)
```

```python
# Simulate delta hedging as seller along the chosen path

cash = american_put_price   # receive premium at t0
shares = 0.0                # shares held by seller (seller holds -delta_long)
node_i = 0
hedge_rows = []

# Initial hedge at t=0
delta0 = delta[0][0]
shares_new = -delta0
trade = shares_new - shares
trade_cost = trade * stock[0][0]   # positive => cash decreases
cash -= trade_cost
shares = shares_new
hedge_rows.append({
    "step": 0, "time": 0.0, "S": stock[0][0], "V": value[0][0],
    "delta_long": delta0, "shares_seller": shares, "trade": trade,
    "trade_cost": trade_cost, "cash_after_trade": cash
})

for n in range(1, N+1):
    # accrue cash
    cash *= math.exp(r * dt)
    move = path[n-1]
    if move == 1:
        node_i += 1
    if n < N:
        S_now = stock[n][node_i]
        V_now = value[n][node_i]
        delta_now = delta[n][node_i]
        shares_new = -delta_now
        trade = shares_new - shares
        trade_cost = trade * S_now
        cash -= trade_cost
        shares = shares_new
        hedge_rows.append({
            "step": n, "time": round(n*dt,8), "S": S_now, "V": V_now,
            "delta_long": delta_now, "shares_seller": shares, "trade": trade,
            "trade_cost": trade_cost, "cash_after_trade": cash
        })
    else:
        # maturity settlement
        S_now = stock[n][node_i]
        payoff = max(K - S_now, 0.0)
        proceeds = shares * S_now
        cash += proceeds
        cash -= payoff
        hedge_rows.append({
            "step": n, "time": round(n*dt,8), "S": S_now, "payoff": payoff,
            "shares_before_unwind": shares, "sell_proceeds": proceeds,
            "cash_after_settlement": cash
        })

hedge_df = pd.DataFrame(hedge_rows).round(8)
print("Hedge evolution (full)")
hedge_df.round(2)

```

```python
# Helper: basis functions for regression (we use simple polynomial basis in (S, A))
def basis_feats(S, A):
    # S and A are 1D arrays of same length
    S = np.asarray(S).reshape(-1,1)
    A = np.asarray(A).reshape(-1,1)
    return np.hstack([np.ones_like(S), S, A, S*A, S**2, A**2])

# Discount factor helper
def discount_factor(r, dt, steps=1):
    return math.exp(-r * dt * steps)
```

```python
# Parameters (same market data as Q25)
S0 = 180.0
K = 182.0
r = 0.02            # continuous annual rate
sigma = 0.25
T = 0.5             # years
N = 25              # time-steps (matching Q26)
dt = T / N

# Monte Carlo settings
M = 40000           # number of simulated paths for LSM (increase for accuracy; 40k is typical)
rng_seed = 2025

# LSM / delta bump
eps_rel = 1e-4      # relative bump for finite difference delta (eps = eps_rel * S)
eps = eps_rel * S0

print("Params:", {"S0":S0,"K":K,"r":r,"sigma":sigma,"T":T,"N":N,"dt":dt,"M":M})
```

```python
np.random.seed(rng_seed)

# Precompute constants
nudt = (r - 0.5 * sigma**2) * dt
sigsqdt = sigma * math.sqrt(dt)

# Simulate lognormal increments
Z = np.random.normal(size=(M, N))  # M paths, N steps
S_paths = np.zeros((M, N+1))
A_paths = np.zeros((M, N+1))  # running arithmetic average including current step
S_paths[:,0] = S0
A_paths[:,0] = S0  # average at time 0 is S0 (convention)

for t in range(1, N+1):
    S_paths[:,t] = S_paths[:,t-1] * np.exp(nudt + sigsqdt * Z[:,t-1])
    # running arithmetic average including the new S_t:
    # A_t = (t * A_{t-1} + S_t) / (t+1)  ; note A_{t-1} stored at index t-1
    A_paths[:,t] = (t * A_paths[:,t-1] + S_paths[:,t]) / (t + 1)

# quick sanity display
display(pd.DataFrame({
    "S0_sample": S_paths[:5,0],
    "S1_sample": S_paths[:5,1],
    "S_last_sample": S_paths[:5,-1],
    "A_last_sample": A_paths[:5,-1]
}))
```

```python
# computing cashflows (exercise times) and estimate continuation value via regression on basis(S_t, A_t).
# Payoff function for a put on the average; we define payoff at time t using the average so far (A_paths[:,t])

def payoff_on_average(A):
    return np.maximum(K - A, 0.0)

# Initialize arrays
cashflow = np.zeros((M, N+1))   # cashflow[t] holds payoff if exercised at time t
exercise_time = np.full(M, N)   # default exercise at maturity
discounts = np.array([math.exp(-r * dt * (t)) for t in range(N+1)])  # discount to time 0

# Terminal payoff at t = N
cashflow[:, N] = payoff_on_average(A_paths[:, N])

# We will work backward from t = N-1 down to 1 (we typically allow exercise at times 1..N)
# At each time t compute continuation value via regression for in-the-money paths, compare to exercise.
# If exercise is optimal, set cashflow at t and mark exercise_time.

# Keep an "alive" mask for paths not yet exercised
alive = np.ones(M, dtype=bool)

# For LSM we need the discounted value of future cashflow for alive paths.
# Start with discounted cashflow from terminal payoff (already set).
discounted_future = cashflow[:, N] * math.exp(-r * dt * (N))  # discounted to time 0, but we will use relative discounted values per step

# We will instead follow standard LSM: at step t, the target is discounted cashflow from next time onward to time t, i.e.
# target = cashflow_at_exercise * exp(-r * dt * (exercise_time - t)), but we compute iteratively.

# To implement more directly: maintain "value" array which contains payoff at exercise if exercised later; initially it's terminal payoff.
exercise_flags = np.zeros(M, dtype=bool)
exercise_payoffs = np.copy(cashflow[:, N])
exercise_times = np.full(M, N)

# Backward induction for LSM
for t in range(N-1, 0, -1):   # allow exercise at 1..N-1 (you can also allow at t=0 if contract allows)
    # For all paths that are still not exercised before t (i.e., exercise_time > t)
    alive = exercise_times > t
    if not np.any(alive):
        continue

    S_t = S_paths[alive, t]
    A_t = A_paths[alive, t]
    immediate_ex = payoff_on_average(A_t)

    # For regression we need the discounted realized cashflows for those paths from future exercise back to time t
    # For each alive path, its realized payoff is exercise_payoffs[path] and realized exercise time is exercise_times[path].
    future_cash = exercise_payoffs[alive] * np.exp(-r * dt * (exercise_times[alive] - t))

    # Only regress on in-the-money paths (immediate_ex > 0)
    itm = immediate_ex > 0
    if np.sum(itm) == 0:
        # no ITM paths at this time
        continue

    X = basis_feats(S_t[itm], A_t[itm])
    Y = future_cash[itm]

    # Fit regression: continuation value ~ basis(S_t, A_t)
    model = LinearRegression(fit_intercept=False)  # basis already contains constant
    model.fit(X, Y)

    # Estimate continuation value for all alive ITM paths
    cont_vals = np.zeros(immediate_ex.shape)
    cont_vals[itm] = model.predict(basis_feats(S_t[itm], A_t[itm]))

    # Exercise decision: if immediate_ex >= continuation, exercise now
    exercise_now = (immediate_ex >= cont_vals) & itm

    # For paths that exercise now, update exercise_payoffs and exercise_times
    idx_alive = np.where(alive)[0]             # indices in original array
    to_exercise_idx = idx_alive[exercise_now]  # original indices that exercise now

    # Update
    exercise_times[to_exercise_idx] = t
    exercise_payoffs[to_exercise_idx] = immediate_ex[exercise_now]

# After backward pass, compute discounted payoff to time 0 and estimate price
discounted_payoffs = exercise_payoffs * np.exp(-r * dt * exercise_times)
price_est = discounted_payoffs.mean()
print(f"LSM estimate for American Asian put price (MC M={M}, N={N}): {price_est:.6f}")#
```

```python
# Note: In the code above we only keep the last model. If you want per-time-step models (to compute deltas later),
# you should store the model for each t where regression occurred. We'll re-run the backward induction and store models.

models = dict()   # models[t] = trained regression object for time t (if regression performed)
exercise_payoffs = np.copy(cashflow[:, N])
exercise_times = np.full(M, N)

for t in range(N-1, 0, -1):
    alive = exercise_times > t
    if not np.any(alive):
        continue
    S_t = S_paths[alive, t]
    A_t = A_paths[alive, t]
    immediate_ex = payoff_on_average(A_t)
    itm = immediate_ex > 0
    if np.sum(itm) == 0:
        continue
    X = basis_feats(S_t[itm], A_t[itm])
    # future cash for those alive paths:
    future_cash = exercise_payoffs[alive] * np.exp(-r * dt * (exercise_times[alive] - t))
    Y = future_cash[itm]
    model = LinearRegression(fit_intercept=False)
    model.fit(X, Y)
    models[t] = model

    # compute continuation value and exercise decision
    cont_vals = np.zeros_like(immediate_ex)
    cont_vals[itm] = model.predict(basis_feats(S_t[itm], A_t[itm]))
    exercise_now = (immediate_ex >= cont_vals) & itm
    idx_alive = np.where(alive)[0]
    to_exercise_idx = idx_alive[exercise_now]
    exercise_times[to_exercise_idx] = t
    exercise_payoffs[to_exercise_idx] = immediate_ex[exercise_now]

# final price
discounted_payoffs = exercise_payoffs * np.exp(-r * dt * exercise_times)
price_est = discounted_payoffs.mean()
print(f"Stored models for times: {sorted(models.keys())}")
print(f"American Asian put price (from stored models): {price_est:.6f}")
```

```python
# Bump S0 up and down, resimulate or reuse initial randomness (pathwise) to compute price difference.
# For consistency, we will resimulate paths with S0 bumped and rerun the LSM procedure.
# For speed you may reuse the same Z but that's a design choice.

def lsm_price_given_S0(S0_new, seed=rng_seed):
    np.random.seed(seed)
    Z = np.random.normal(size=(M, N))
    S_p = np.zeros((M, N+1))
    A_p = np.zeros((M, N+1))
    S_p[:,0] = S0_new
    A_p[:,0] = S0_new
    for t in range(1, N+1):
        S_p[:,t] = S_p[:,t-1] * np.exp((r-0.5*sigma**2)*dt + sigma*math.sqrt(dt) * Z[:,t-1])
        A_p[:,t] = (t * A_p[:,t-1] + S_p[:,t]) / (t + 1)
    # Run LSM with stored models structure: we will recompute exercise via stored models
    exercise_pay = np.copy(np.maximum(K - A_p[:,N], 0.0))
    exercise_times_loc = np.full(M, N)
    # backward using stored models
    for t in range(N-1, 0, -1):
        alive = exercise_times_loc > t
        if not np.any(alive):
            continue
        S_t = S_p[alive, t]
        A_t = A_p[alive, t]
        imm = np.maximum(K - A_t, 0.0)
        if t in models:
            model = models[t]
            cont = model.predict(basis_feats(S_t, A_t))
            exercise_now = imm >= cont
            idx_alive = np.where(alive)[0]
            to_ex = idx_alive[exercise_now]
            exercise_times_loc[to_ex] = t
            exercise_pay[to_ex] = imm[exercise_now]
        else:
            # if no model because no ITM training paths, skip
            continue
    discounted = exercise_pay * np.exp(-r*dt*exercise_times_loc)
    return discounted.mean()

# central difference
h = eps
price_plus = lsm_price_given_S0(S0 + h)
price_minus = lsm_price_given_S0(S0 - h)
delta0 = (price_plus - price_minus) / (2*h)
print(f"Delta at t=0 (approx): {delta0:.6f}")
```

```python
# Hedging along one chosen path using delta from regression surface
# We'll pick one realized path from the original S_paths (e.g., path index 0 for simplicity)
path_idx = 0
S_path = S_paths[path_idx].copy()
A_path = A_paths[path_idx].copy()

# Seller receives premium at t0
premium = price_est
cash = premium
shares = 0.0

# Function to compute delta_long at time t given (S_t, A_t) using stored model
def delta_from_models(t, S_t, A_t, h=eps):
    # If there's a model at time t, we can approximate derivative of continuation w.r.t S_t via finite diff
    # When S_t is bumped by +h, the average A_t changes by +h/(t+1) (since A_t = (t*A_prev + S_t)/(t+1)).
    if t not in models:
        # fallback: compute central difference by bumping S and re-evaluating entire LSM pricing locally (expensive)
        # Here we return 0.0 as a conservative fallback
        return 0.0
    model = models[t]
    A_up = A_t + h/(t+1)
    A_down = A_t - h/(t+1)
    cont_up = model.predict(basis_feats(np.array([S_t + h]), np.array([A_up])))[0]
    cont_down = model.predict(basis_feats(np.array([S_t - h]), np.array([A_down])))[0]
    # total option value at time t is max(immediate_exercise, continuation). But delta should reflect continuation derivative,
    # and when immediate exercise is optimal the delta is simply derivative of intrinsic w.r.t S (which is 0 for put on average, because intrinsic uses A_t not S_t — subtle).
    delta = (cont_up - cont_down) / (2*h)
    return delta

# initial delta and initial hedge
delta_long0 = delta0  # from earlier global bump; alternatively compute via delta_from_models(0,...)
shares_new = -delta_long0
trade = shares_new - shares
trade_cost = trade * S_path[0]
cash -= trade_cost
shares = shares_new

hedge_history = []
hedge_history.append({
    "step": 0, "time": 0.0, "S": S_path[0], "A": A_path[0], "delta_long": delta_long0,
    "shares_seller": shares, "trade": trade, "trade_cost": trade_cost, "cash": cash
})

# simulate rebalancing along the path for t=1..N-1 (if option not exercised earlier)
exercise_happened = False
for t in range(1, N+1):
    # accrue cash
    cash *= math.exp(r * dt)
    S_t = S_path[t]
    A_t = A_path[t]
    imm = max(K - A_t, 0.0)
    # check if at this t the holder would (according to stored models) exercise
    # Using stored model to estimate continuation
    if t in models:
        cont = models[t].predict(basis_feats(np.array([S_t]), np.array([A_t])))[0]
    else:
        cont = 0.0
    if imm >= cont:
        # early exercise happens at t
        payoff = imm
        # seller must pay payoff and unwind hedge by selling shares at S_t
        proceeds = shares * S_t
        cash += proceeds
        cash -= payoff
        hedge_history.append({
            "step": t, "time": t*dt, "S": S_t, "A": A_t, "exercise": True,
            "payoff": payoff, "shares_unwound": shares, "proceeds": proceeds, "cash": cash
        })
        exercise_happened = True
        break
    else:
        # continue: compute delta from models and rebalance
        # use delta_from_models for local delta
        delta_long_t = delta_from_models(t, S_t, A_t, h=eps)
        shares_new = -delta_long_t
        trade = shares_new - shares
        trade_cost = trade * S_t
        cash -= trade_cost
        shares = shares_new
        hedge_history.append({
            "step": t, "time": t*dt, "S": S_t, "A": A_t, "delta_long": delta_long_t,
            "shares_seller": shares, "trade": trade, "trade_cost": trade_cost, "cash": cash
        })

# If no early exercise until maturity, settle at maturity
if not exercise_happened:
    S_T = S_path[-1]
    A_T = A_path[-1]
    payoff = max(K - A_T, 0.0)
    proceeds = shares * S_T
    cash += proceeds
    cash -= payoff
    hedge_history.append({
        "step": N, "time": N*dt, "S": S_T, "A": A_T, "exercise": False,
        "payoff": payoff, "shares_unwound": shares, "proceeds": proceeds, "cash": cash
    })

hedge_df = pd.DataFrame(hedge_history)
display(hedge_df.head(20))
print("Final cash after hedging (seller) on chosen path:", cash)
```

The hedge-evolution results for the Asian ATM Put in Q27 show a much smoother and more stable delta process compared to the American Put in Q25. Because the Asian option payoff depends on the average price rather than the spot price, the delta changes more gradually at each step, resulting in smaller hedge rebalancing trades and more stable cash-account movements. There is no early-exercise feature for the Asian Put, and therefore no sharp jumps in delta or sudden changes in hedge position that typically occur near the early-exercise boundary of an American option.

At maturity, the hedger unwinds the delta position, collects proceeds from selling the hedge, and pays out the Asian Put payoff based on the final average price. The final cash balance of approximately 12.74 indicates a positive hedging outcome, which contrasts with the negative hedging error -11.01 observed for the American Put in Q25. This is consistent with financial theory: Asian options generally have lower hedging error because the averaging mechanism smooths price volatility, while American options exhibit larger hedging error due to the early-exercise feature, which makes replication more difficult.
